```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import KFold
from imblearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, precision_score, recall_score, average_precision_score
```

In this notebook, I will be using classification algorithms to recommend a list of football games to rewatch. To maximize the precision and recall of the model, I will compare the results of three classification algorithms: Logistic Regression, Decision Tree, and Random Forest. I will then evaluate the accuracy and recall by comparing the precision, recall, and F1 scores of each model.

```python
# Read in data
df = pd.read_csv('EPL_model_ready.csv', index_col=False)
df
```

I will first split the data into the independent and dependent variables. Note that I will not be using all the columns in the dataframe as columns several columns ("Comeback", "Total Goals", etc.) are derived from other columns in the table (FTR, HTR, etc.). I will be removing all non-numerical columns.

```python
features = df.drop(columns=['Date','HomeTeam','AwayTeam','Season','DateTime'])
features
```

### Feature Selection

In order to reduce overfitting and training time, I will only train the model with features that correlate the most with the dependent variable. To do so, I will plot a correlation heatmap and take the top five features.

```python
plt.figure(figsize=(20,10))
corrmat = features.corr()
corr = sns.heatmap(corrmat, cmap='coolwarm', annot=True)
```

According to the heatmap, the top five features are: 

- Total Goals
- FTHG (Full Time Home Goals)
- FTAG (Full Time Away Goals)
- Rival Game
- Comeback

However, we also see that Total Goals is very highly correlated to FTAG and FTHG. This is because Total Goals is the sum of FTAG and FTHG. To ensure collinearity, we will not consider FTAG and FTHG. We take the next 4 most highly correlated variables, resulting in the following features:

- Total Goals
- Rival Game
- Comeback
- Popularity Score
- ~~HTAG (Half Time Away Goals): also highly correlated with Total Goals~~
- ~~HTHG (Half Time Hway Goals): also highly correlated with Total Goals~~
- ~~Goal Difference: also highly correlated with Total Goals~~
- HTR (Half Time Result)

```python
# Split data into dependent variable and independent variables based on explaination above
X = features[['Total Goals','Rival Game','Comeback','Popularity Score','HTR']]
X = X.to_numpy()
y = features[['Best Game']]
y = y.to_numpy()
print(X)
```

### Balancing classes using SMOTE

As it stands, there are 100 samples that are classified as a "Best Game" while 8640 samples are not. Since there is a clear imbalance between the number of samples in each class, I will need to either oversample data from the minority class (the 'Best Games' class), or undersample data from the majority class to compensate. I have chosen to oversample from the minority class using SMOTE (Synthetic Minority Oversampling TEchnique) because if I undersample the classifier may not have enough data (only 200 samples) to classify the samples with high precision and recall. 

### Pipelines and KFold Cross Validation

I will be using Pipelines to enable the cross validation the process of applying SMOTE and the classification model.

I will be using the KFolds cross-validator to split the data into train/test splits. As per the Pareto Principle, I will be using a 80/20 train-test split, so the data will be split into 5 consecutive folds. 

```python
kf = KFold(n_splits=5)
f1_lr = [] # To store f1 scores of logistic regression
f1_dt = [] # To store f1 scores of decision tree
f1_rf = [] # To store f1 scores of random forest
pre_lr = [] # To store precision scores of logistic regression
pre_dt = [] # To store precision scores of decision tree
pre_rf = [] # To store precision scores of random forest
re_lr = [] # To store recall scores of logistic regression
re_dt = [] # To store recall scores of decision tree
re_rf = [] # To store recall scores of random forest

# Pipelines
smt = SMOTE(random_state=42)
lr = LogisticRegression()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
pipeline_lr = Pipeline([('smt', smt), ('lr', lr)])
pipeline_dt = Pipeline([('smt', smt), ('lr', dt)])
pipeline_rf = Pipeline([('smt', smt), ('lr', rf)])
pipelines = [pipeline_lr, pipeline_dt, pipeline_rf]
```

```python
for train_index, test_index in kf.split(X,y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    for pipeline in pipelines:
        model = pipeline.fit(X_train, y_train)
        prediction = model.predict(X_test)
        if pipeline == pipeline_lr:
            f1_lr.append(f1_score(y_test, prediction))
            pre_lr.append(precision_score(y_test, prediction))
            re_lr.append(recall_score(y_test, prediction))
        elif pipeline == pipeline_dt:
            f1_dt.append(f1_score(y_test, prediction))
            pre_dt.append(precision_score(y_test, prediction))
            re_dt.append(recall_score(y_test, prediction))
        else:
            f1_rf.append(f1_score(y_test, prediction))
            pre_rf.append(precision_score(y_test, prediction))
            re_rf.append(recall_score(y_test, prediction))
            
print("Logistic Regression F1: {}".format(np.mean(f1_lr)))
print("Logistic Regression precision: {}".format(np.mean(pre_lr)))
print("Logistic Regression recall: {}".format(np.mean(re_lr)))
print('\n')
print("Decision Tree F1: {}".format(np.mean(f1_dt)))
print("Decision Tree precision: {}".format(np.mean(pre_dt)))
print("Decision Tree recall: {}".format(np.mean(re_dt)))
print('\n')
print("Random Forest F1: {}".format(np.mean(f1_rf)))
print("Random Forest precision: {}".format(np.mean(pre_rf)))
print("Random Forest recall: {}".format(np.mean(re_rf)))
```
